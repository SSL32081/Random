```python
import ipywidgets

from ipywidgets.embed import embed_minimal_html
```

```python
rand1 = np.random.normal(0, 4, 100)
rand2 = np.random.normal(1, 3, 100)

def plot_noise_xy(wx, wy):
    fig, ax = plt.subplots(1,1, figsize=(10, 10))    
    ax.scatter(wx*rand1, wy*rand2)
    return 
    
ipywidgets.interact(plot_noise_xy, wx=range(5), wy=range(5))
```
